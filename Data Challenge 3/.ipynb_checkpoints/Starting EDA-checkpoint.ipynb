# Exploratory Data Analysis

We first explore the question of whether the **company should shift from a single pricew of 39 dollars to a single price of 59 dollars per software license**. 

Since software is a digital good, there is minimal cost associated with selling additional licenses. The possible exception is any ongoing support costs that are associated with each license. This cost may be small and scale sub-linearly in the number of licenses sold.



```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


from IPython.display import set_matplotlib_formats
set_matplotlib_formats('png', 'pdf')

%matplotlib inline

test_results = pd.read_csv('test_results.csv')
user_table = pd.read_csv('user_table.csv')
```

```python
test_results.head()
```

```python
user_table.head()
```

```python
test_results['converted'].value_counts()
# conversions are very very rare
```

```python
test_results['price'].value_counts()
```

```python
# Let's check the fraction of observed conversions at each price point

control_cohort = test_results[test_results.price == 39]
alternative_cohort = test_results[test_results.price == 59]

print('Separate into control and alternative treatment tables: ')
print('Control: ' + str(control_cohort.shape))
print('Alternative Higher Price: ' + str(alternative_cohort.shape))

control_conversions = control_cohort[control_cohort.converted == 1]
alternative_conversions = alternative_cohort[alternative_cohort.converted == 1]

print(' ')
print('What about conversions only? ')
print('Control: ' + str(control_conversions.shape))
print('Alternative Higher Price: ' + str(alternative_conversions.shape))
```

```python
control_conversion_rate = len(control_conversions)/len(control_cohort)
print('Control conversion rate: ' + str(control_conversion_rate))

alternative_conversion_rate = len(alternative_conversions)/len(alternative_cohort)
print('Alternative conversion rate: ' + str(alternative_conversion_rate))

ratio_of_conversion_rates = alternative_conversion_rate/control_conversion_rate
print(' ')
print('Ratio of conversion rates: ' + str(ratio_of_conversion_rates))
```

```python
# Should we sell at $39 or $59?
# Suppose that we do set a single fixed price and show this to 100,000 people
# Assuming that the control and treatment group were randomly specified,
# we estimate the following revenue for the two options:

revenue_control_price = 39*control_conversion_rate*100000
print('Estimated control revenue expected from 100,000 Webpage-arrivals: ' 
      + str(round(revenue_control_price)) + ' dollars')

revenue_alternative_price = 59*alternative_conversion_rate*100000
print('Estimated alternative revenue expected from 100,000 Webpage-arrivals: ' 
      + str(round(revenue_alternative_price)) + ' dollars')

print(' ')
print('So, we estimate that revenue will be ' + str(round((revenue_alternative_price/revenue_control_price-1)*100)) 
      + '% higher if we sell at the higher price (while maintaining the mix of current traffic sources).')

print(' ')
print('Further, the total of any on-going costs of supporting liscences may be less at the higher price since about '+
      str(round(100*(1-ratio_of_conversion_rates)))+ '% fewer liscenes will be sold.')
```

**Discussion:** Based on this first-round of analysis, it seems promising to sell the software at the higher price. However, we have **not yet considered the variance in conversion rates for the two treatments** (we have used only point estimates for the conversion rate in each group).

So, let's now consider variance. We will continue to **focus on the expected revenue from 100,000 webpage-arrivals**, but we will now use resampling to describe the variance around these estimated revenue figures.

```python
# We will repeatedly resample populations of 100,000 from each group
# to build an estimate of the distribution of profits at both price levels

resampled_control_revenues = []

for index in range(1000):
    resampled_control_cohort = control_cohort.sample(100000, replace = True)
    resampled_control_conversions = resampled_control_cohort[resampled_control_cohort.converted == 1]
    resampled_control_conversion_rate = len(resampled_control_conversions)/len(resampled_control_cohort)
    resampled_revenue_control_price = 39*resampled_control_conversion_rate*100000
    resampled_control_revenues.append(resampled_revenue_control_price)
```

```python
plt.rcParams["figure.figsize"] = (8,6)
plt.hist(resampled_control_revenues)
plt.title('Estimated Revenue per 100,000 Webpage-arrivals at $39 Price')
plt.ylabel('Frequency (out of 1,000)')
plt.xlabel('Dollars of Revenue per 100,000 Webpage-arrivals')

resampled_control_revenues = pd.Series(resampled_control_revenues)

plt.axvline(x = resampled_control_revenues.mean(),color ='black')

print(resampled_control_revenues.mean())
print('Lower-bound of confidence interval: ' + str(round(resampled_control_revenues.quantile(0.025))))
print('Upper-bound of confidence interval: ' + str(round(resampled_control_revenues.quantile(0.975))))
```

```python
# We will repeatedly resample populations of 100,000 from each group
# to build an estimate of the distribution of profits at both price levels

resampled_alternative_revenues = []

for index in range(1000):
    resampled_alternative_cohort = alternative_cohort.sample(100000, replace = True)
    resampled_alternative_conversions = resampled_alternative_cohort[resampled_alternative_cohort.converted == 1]
    resampled_alternative_conversion_rate = len(resampled_alternative_conversions)/len(resampled_alternative_cohort)
    resampled_revenue_alternative_price = 59*resampled_alternative_conversion_rate*100000
    resampled_alternative_revenues.append(resampled_revenue_alternative_price)
```

```python
plt.rcParams["figure.figsize"] = (8,6)
plt.hist(resampled_alternative_revenues)
plt.title('Estimated Revenue per 100,000 Webpage-arrivals at $59 Price')
plt.ylabel('Frequency (out of 1,000)')
plt.xlabel('Dollars of Revenue per 100,000 Webpage-arrivals')

resampled_alternative_revenues = pd.Series(resampled_alternative_revenues)

plt.axvline(x = resampled_alternative_revenues.mean(),color ='black')

print(resampled_alternative_revenues.mean())
print('Lower-bound of confidence interval: ' + str(round(resampled_alternative_revenues.quantile(0.025))))
print('Upper-bound of confidence interval: ' + str(round(resampled_alternative_revenues.quantile(0.975))))
```

```python
plt.rcParams["figure.figsize"] = (12,7)


plt.hist(resampled_control_revenues, alpha=0.3, color = 'grey', label='Estimated Revenue at $39 Price', bins=np.arange(0, 120000, 2000))
plt.hist(resampled_alternative_revenues,alpha=0.6, color = 'green', label='Estimated Revenue at $59 Price',bins=np.arange(0, 120000, 2000))
plt.axvline(x = resampled_control_revenues.mean(),color ='grey')
plt.axvline(x = resampled_alternative_revenues.mean(),color ='green')
plt.ylabel('Frequency (out of 1,000)')
plt.xlabel('Dollars of Revenue per 100,000 Webpage-arrivals')
plt.xlim(0,120000)

plt.legend(loc='upper left')
plt.title('Changing to $59 Price Will Improve Revenues (Expected Revenue boost of 18%)')
```

```python
plt.rcParams["figure.figsize"] = (10,6)

rand_differences = []

for index in list(range(len(resampled_control_revenues))):
    rand_differences.append(100*(-resampled_control_revenues[index]+resampled_alternative_revenues[index])/resampled_control_revenues.mean())

rand_differences = pd.Series(rand_differences)    
plt.hist(rand_differences, alpha=0.5,color = "orange")
plt.axvline(x = rand_differences.mean(),color ='orange')
plt.ylabel('Frequency (out of 1,000)')
plt.xlabel('Predicted Percentage Increase in Revenue')
plt.xlim(0,35)

#plt.legend(loc='upper right')
plt.title('Changing to $59 Price Will Improve Revenues (Expected Revenue boost of 18%)')

print('95% of predictions fall between ' + str(round(rand_differences.quantile(0.025))) +'% increase and ' 
     + str(round(rand_differences.quantile(0.975))) + '% increase in Revenue.')
```

The previous two graphics appear to indicate that **switching to the higher sale price** (across all sales) could substantially boost revenue (based on the experimental evidence, we expect a **revenue boost between 11% and 25%, with a mean expected revenue increase of 18%**).

The analysis above is based on **maintaining the same mix of traffic sources** as the company has expereinced over the testing period. Notably, some of these traffic sources are hard for the company to directly control, like placement in non-sponsored search results (seo catagories for various browsers).

Note: if there are any **ongoing support costs that the company has to cover after the initial sale**, then we expect profits to be increased above the boost predicted by considering revenue only (as mentioned above, at the increased price point, roughly 22% fewer licenses are sold).

*Second Note: since the real sample sizes are even larger than those we've used for resampling, we expect that the distributions above actually over-estimate the variability of the revenue estimates. The true distributions of the sample statistic should be even more concentrated/narrow. Given that, even at sample sizes of only 100,000 we see a high amount of separation between the two distributions, we are convinced that the true distributions are very-well-separated.* 

## Verify Unbiased Study Design

While the figures above appeared very convincing, for due dilligence, before reporting these results we will **check to verify that our control vs. alternative treatments were truly distributed uniformly across all platforms**. In particular, if the alternative treatment was disproportionately applied to a platform with unusually high conversion rates (aka, there was a bias in how treatment status was assigned) then our conclusions about the efficacy of switching to the higher price could be faulty.

```python
test_results.head()
```

```python
test_results['source'].value_counts()
```

```python
# Was the treatment assigned randomly across the advertising platforms?

check_study_design = test_results[['source','test']]
print(check_study_design.groupby(['source']).mean())

grouped_by_platform = check_study_design.groupby(['source']).mean()

x = grouped_by_platform.index.tolist()

height = grouped_by_platform['test']

plt.barh(x, height)
plt.title('Approximately Uniform Treatment Assignment Across Traffic Source')
```

```python
# Was the treatment assigned randomly across the device types?

check_study_design = test_results[['device','test']]
check_study_design.groupby(['device']).mean()
```

The study design looks quite uniform across both of these splits (traffic source and device type). Since **it doesn't appear that there were obvious biases in how treatments were assigned**, we are comfortable affirming our finding that revenue will increase after a shift to the higher price point.

After switching to the new price of 59 dollars, is there evidence that the **company should pay different amounts for click throughs** from different ad placement services?



```python
ads_set = ['ads-google','ads_facebook', 'ads_other', 'ads-bing','ads-yahoo']
ad_traffic_indic = []
for index in list(range(len(test_results))):
    if test_results['source'][index] in ads_set:
        ad_traffic_indic.append(True)
    else:
        ad_traffic_indic.append(False)
        
ad_traffic_only = test_results[ad_traffic_indic]
```

```python
ad_traffic_only['source'].value_counts()
```

```python
ad_traffic_only.groupby(['source']).mean()
```

```python
dict_of_randomized_conversion_rates = {}

for ad_type in ads_set:
    
    print(ad_type)
    
    limit_to_particular_ads = ad_traffic_only[ad_traffic_only.source == ad_type]
    
    resampled_alternative_revenues = []
    
    sample_size = round(.9*len(limit_to_particular_ads))

    for index in range(1000):
        resampled_alternative_cohort = limit_to_particular_ads.sample(sample_size, replace = True)
        resampled_alternative_conversions = resampled_alternative_cohort[resampled_alternative_cohort.converted == 1]
        resampled_alternative_conversion_rate = len(resampled_alternative_conversions)/len(resampled_alternative_cohort)
        resampled_revenue_alternative_price = 59*resampled_alternative_conversion_rate
        resampled_alternative_revenues.append(resampled_revenue_alternative_price)
        
    dict_of_randomized_conversion_rates[ad_type] = resampled_alternative_revenues
    
```

```python
plt.rcParams["figure.figsize"] = (12,7)


plt.hist(dict_of_randomized_conversion_rates['ads-bing'], alpha=0.3, color = 'grey', label='Bing Ads',bins=np.arange(0, 2, 0.05))
plt.hist(dict_of_randomized_conversion_rates['ads-yahoo'], alpha=0.6, color = 'grey', label='Yahoo Ads',bins=np.arange(0, 2, 0.05))
plt.hist(dict_of_randomized_conversion_rates['ads_other'], alpha=0.6, color = 'black', label='Other Ads',bins=np.arange(0, 2, 0.05))
plt.hist(dict_of_randomized_conversion_rates['ads_facebook'], alpha=0.5, color = 'green', label='Facebook Ads',bins=np.arange(0, 2, 0.05))
plt.hist(dict_of_randomized_conversion_rates['ads-google'], alpha=0.4, color = 'blue', label='Google Ads',bins=np.arange(0, 2, 0.05))

plt.ylabel('Frequency (out of 1,000)')
plt.xlabel('Expected revenue per click-through (dollars)')
plt.xlim(0,2)

plt.legend(loc='upper left')
plt.title('Revenue-per-clickthrough is highest for Google and Facebook Ads')
```

```python
# Let's give some confidence intervals about the revenue values 
# associated with click throughs from each ad type:

for ad_type in ads_set:
    print(ad_type)
    series_of_realizations = pd.Series(dict_of_randomized_conversion_rates[ad_type])
    
    lower_bound = series_of_realizations.quantile(0.025)
    upper_bound = series_of_realizations.quantile(0.975)
    mean_rev = series_of_realizations.mean()
    print('Estimated Revenue-per-clickthrough: ' + str(mean_rev))
    print('Revenue-per-clickthrough Confidence Interval: (' + str(round(lower_bound,3))+','+ str(round(upper_bound,3)) + ')')
    print(' ')
```

We find strong evidence that the **expected revenue varies from different types of clickthrough traffic**. For example, if the company can choose to pay either 60 cents per clickthrough for a Bing ad vs 80 cents per clickthrough for a Google Ad, the **revenue margin would be stronger for Google ads even though the Google clickthroughs are more expensive**.
