# Evidence indicating different click-through rates

Suppose that you have run a simple A/B test with results as follows:

5000 displays of A treatment: 1100 clicks (aka 22% click rate)

5000 displays of B treatment: 1300 clicks (aka 26% click rate)

These numbers are not identical, but could these be two realizations from a common distribution? 


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
realizations_from_trial = []

for trial in range(1000):
    heads = np.random.binomial(5000, 0.24)
    realizations_from_trial.append(heads)
    
plt.hist(realizations_from_trial, bins = 20)
plt.title('1000 Trials of 5000 fair coin flips with common probability of 24%')
plt.xlabel('Heads observed from 5000 samples')
plt.ylabel('Frequency')

realizations = pd.Series(realizations_from_trial)
plt.axvline(x = realizations.mean(),color ='black')
```
